```python
import matplotlib.pyplot as plt
import numpy as np

def generate_wave(amplitude, frequency, duration, sample_rate):
    t = np.linspace(0, duration, int(sample_rate * duration), endpoint=False)
    wave = amplitude * np.sin(2 * np.pi * frequency * t)
    return t, wave

# :     440  ( )
amplitude = 1
frequency = 440
duration = 1  # 1 
sample_rate = 44100

t, wave = generate_wave(amplitude, frequency, duration, sample_rate)
plt.plot(t[:1000], wave[:1000])  #   1000 
plt.xlabel(" ()")
plt.ylabel("")
plt.title(" ")
plt.show()
```

```python
import numpy as np
import sounddevice as sd

def play_sine_wave(frequency, duration, amplitude=0.5, sample_rate=44100):
    """
        .
    
    :param frequency:    
    :param duration:    
    :param amplitude:   ( 0  1)
    :param sample_rate:   (  44100 )
    """
    t = np.linspace(0, duration, int(sample_rate * duration), endpoint=False)
    wave = amplitude * np.sin(2 * np.pi * frequency * t)
    sd.play(wave, samplerate=sample_rate)
    sd.wait()  #   

# :    (440 )  2 
play_sine_wave(440, 2)
```
